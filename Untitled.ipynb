```python
# -*- coding: utf-8 -*-
#相应模块的导入
import re
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns 
from IPython.display import display

plt.style.use("fivethirtyeight")
sns.set_style({'font.sans-serif':['simhei','Arial']})

'''
数据处理
'''
#导入链家二手房数据并添加列名
lianjia_df = pd.read_csv("lianjia.csv",header=None)
lianjia_df.columns=['TotalPrice','PerPrice','Floor','Message','Popularity']
#数据备份
df = lianjia_df.copy()
display(df.head(5))

#对Floor,Message和Popularity三列分别分割字符串，然后与df合并


#以‘|’进行分割，分割出来的内容按所规定的列名成列....Message
split1=df['Message'].str.split('|',expand=True)
split1.columns=['Region','Layout','Area','Direction','Decoration','Elevator']
#display(split1)
#将分隔开的字符与原数据表合并........................Message
if type(split1)==type(df):
    df=pd.concat([df.drop('Message',axis=1),split1],axis=1)
    #display(df.head(5))
else:
    print("--------------类型不同，不可合并！----------------")
#将floor栏信息分开...................................Floor
split2=df['Floor'].str.split('-',expand=True)
split2.columns=['storey','location']
#display(split2)
#将分隔开的字符与原数据表合并.........................Floor
df=pd.concat([df.drop('Floor',axis=1),split2],axis=1)
#display(df.head(5))
#将popularity栏信息分隔开.............................Popularity
split3=df['Popularity'].str.split('/',expand=True)
split3.columns=['attention','view_times','release_date']
#display(split3)
#将分隔开的字符与原数据表合并..........................Popularity
df=pd.concat([df.drop('Popularity',axis=1),split3],axis=1)
# display(df.head(5))
#将storey再细分为三列，并与原数据表合并
split4=df['storey'].str.split(')',expand=True)
split4.columns=['floor','year_type']
split5=split4['floor'].str.split('(',expand=True)
split5.columns=['loc_floor','total_floor']
split6=split4['year_type'].str.split('建',expand=True)
split6.columns=['year','type']
df=pd.concat([df.drop('storey',axis=1),split5],axis=1)
df=pd.concat([df,split6],axis=1)
display(df.head(5))


#价格，面积等列只留下数字
df['Area']=df['Area'].str.replace('平米','')
df['PerPrice']=df['PerPrice'].str.replace('[单价/平米元]','')
df['attention']=df['attention'].str.replace('人关注','')
df['view_times']=df['view_times'].str.replace('[共次带看]','')
df['total_floor']=df['total_floor'].str.replace('[共层]','')
display(df.head(5))
```
